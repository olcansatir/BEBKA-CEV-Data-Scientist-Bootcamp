**TASK-3B: Please provide an analysis (mean and std) for the volumes of hippocampus with respect to the level of AD(Alzheimer's Disease).**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data_labels = pd.read_csv("CDR_Values_y_train.csv")
data_all = pd.read_csv("MRI_and_CDRinfo_Values_X_train.csv")
```

```python
data_all
```

```python
data_labels
```

```python
data_labels['HIP_VOL'] = data_all['HIPPOVOL']
```

```python
data_all["HIPPOVOL"]
```

```python
print("MEAN")
print(data_labels.groupby('CDRGLOB')[['HIP_VOL']].mean())
print("\n\n")
print("Standard Deviation")
print("\n")
print(data_labels.groupby('CDRGLOB')[['HIP_VOL']].std())
```

```python
data_labels.groupby('CDRGLOB').agg({ 'HIP_VOL': ['mean','std']}).plot(kind='bar',grid=True)
plt.title("Means and Standard Deviations")
plt.ylabel("Value")
plt.xlabel("AD(Alzheimer's Disease)")
```
