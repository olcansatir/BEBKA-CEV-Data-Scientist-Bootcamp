```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
data = pd.read_excel("Temporary_data3_Left_Right_Copy.xlsx")
```

```python
pd.set_option('display.max_columns', None)
data
```

```python
x = data.iloc[:,38:39] 
y = data.iloc[:,32:33]
df1  = data.iloc[:,:].div(data.NACCBRNV, axis=0) #divide
df2 = df1.iloc[:,:].multiply(df1.NACCAGEB, axis = 0) #multiply
pd.set_option('display.max_columns', None)
print(df1)
print(df2)
```

**TASK-9B: Please apply Gaussian Naive Bayes algorithm on the numeric values (after the transformation) to predict the CDRGLOB values (class-based). Train-test split, train by grid search (k=5) and optimize for the parameter values**

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
y = data["CDRGLOB"]
X = df2

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,
                                  test_size=0.2)

gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
accuracy_score(y_test, y_pred)

params_NB = {'var_smoothing': np.logspace(0,-9, num=100)}
grid = GridSearchCV(estimator=gnb, 
                 param_grid=params_NB, 
                 cv=5) 
grid.fit(X_train, y_train);
grid.best_params_
gnb = grid.best_estimator_
my_predictions = gnb.fit(X_train,y_train).predict(X_test)
print(classification_report(y_test, my_predictions))
     
```

**TASK-9F: Please find and apply the classification versions of the following algorithms (Linear, Ridge and Lasso) with the hyperparameter optimization.**

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
model = LinearRegression()
model.get_params()
linear_p = {"copy_X": [True, False],
             "fit_intercept": [True, False],
             "positive": [True, False],
             "n_jobs": np.arange(5)}
l_gr = GridSearchCV(LinearRegression(), linear_p, cv=5)
l_gr.fit(X_train, y_train)
print("Score",l_gr.best_score_)
```

```python
model = Ridge()
model.get_params()
```

```python
ridge_p = {"copy_X": [True, False],
             "fit_intercept": [True, False],
             "positive": [True, False],
             "alpha": np.arange(5),
           "max_iter": np.arange(5)}
r_gr = GridSearchCV(model, ridge_p, cv=5)
r_gr.fit(X_train, y_train)
print("Score",r_gr.best_score_)
```

```python
model = Lasso()
model.get_params()
```

```python
lasso_p = {"copy_X": [True, False],
             "fit_intercept": [True, False],
             "positive": [True, False],
             "alpha": np.arange(5),
           "max_iter": np.arange(5),
           "precompute": [True, False]}
l_gr = GridSearchCV(model,lasso_p, cv=5)
l_gr.fit(X_train, y_train)
print("Score",l_gr.best_score_)
```
